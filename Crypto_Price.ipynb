## Preprocessing

```python
btc_df = btc_df.loc['20150808235959':]
```

```python
btc_d1 = []
btc_d2 = []

for i in range(1, len(btc_df)):
    if btc_df['Close'][i-1] < btc_df['Close'][i]:
        btc_d1.append(1)
    else:
        btc_d1.append(0)
    btc_d2.append(btc_df['Close'][i] - btc_df['Close'][i-1])
```

```python
eth_d1 = []
eth_d2 = []

for i in range(1, len(eth_df)):
    if eth_df['Close'][i-1] < eth_df['Close'][i]:
        eth_d1.append(1)
    else:
        eth_d1.append(0)
    
    eth_d2.append(eth_df['Close'][i] - eth_df['Close'][i-1])
```

```python
d = {'Date': btc_df['Date'][:-1], '+1D BTC': btc_d1, '+1D ETH': eth_d1, 'Price Diff BTC':btc_d2, 'Price Diff ETH':eth_d2, 'BTC Volume':btc_df.iloc[:-1, 8], 'ETH Volume':eth_df.iloc[:-1, 8]}
df = pd.DataFrame(data= d)
df = df.drop('Date', axis=1)
```
